```python
X_red = df_red.drop('quality', axis=1)
y_red = df_red['quality']

X_white = df_white.drop('quality', axis=1)
y_white = df_white['quality']
```

### 2. Train-test split with `test_size=0.2`, `random_state=42`, `stratify=y`

```python
from sklearn.model_selection import train_test_split

# Red wine train-test split
X_train_red, X_test_red, y_train_red, y_test_red = train_test_split(X_red, y_red, test_size=0.2, random_state=42, stratify=y_red)

# White wine train-test split
X_train_white, X_test_white, y_train_white, y_test_white = train_test_split(X_white, y_white, test_size=0.2, random_state=42, stratify=y_white)

print("Red Wine Dataset Shapes:")
print(f"X_train_red: {X_train_red.shape}, y_train_red: {y_train_red.shape}")
print(f"X_test_red: {X_test_red.shape}, y_test_red: {y_test_red.shape}")

print("\nWhite Wine Dataset Shapes:")
print(f"X_train_white: {X_train_white.shape}, y_train_white: {y_train_white.shape}")
print(f"X_test_white: {X_test_white.shape}, y_test_white: {y_test_white.shape}")
```

### Task 4 — Model A (Baseline): Logistic Regression (Red vs White)

#### 1. Build and Train Logistic Regression Pipeline for Red Wine

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Create a Pipeline for red wine
pipeline_red = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=5000))
])

# Fit the pipeline to the red wine training data
pipeline_red.fit(X_train_red, y_train_red)

print("Logistic Regression pipeline for red wine trained successfully.")
```

#### 2. Evaluate Red Wine Model Performance

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score

# Evaluate red wine model on the test set
y_pred_red = pipeline_red.predict(X_test_red)

print("--- Red Wine Model Evaluation ---")
accuracy_red = accuracy_score(y_test_red, y_pred_red)
print(f"Accuracy of Red Wine: {accuracy_red:.4f}")

confusion_red = confusion_matrix(y_test_red, y_pred_red)
print("\nConfusion Matrix of Red Wine:\n", confusion_red)

classification_report_red = classification_report(y_test_red, y_pred_red, zero_division=0)
print("\nClassification Report of Red Wine:\n", classification_report_red)

macro_f1_red = f1_score(y_test_red, y_pred_red, average='macro', zero_division=0)
print(f"Macro F1-score for Red Wine: {macro_f1_red:.4f}")
#print(f"Accuracy: {accuracy_score(y_test_red, y_pred_red):.4f}")
#print("\nConfusion Matrix:\n", confusion_matrix(y_test_red, y_pred_red))
#print("\nClassification Report (macro F1-score emphasized):\n", classification_report(y_test_red, y_pred_red, zero_division=0))
```

#### 3. Build and Train Logistic Regression Pipeline for White Wine

```python
# Create a Pipeline for white wine
pipeline_white = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=5000))
])

# Fit the pipeline to the white wine training data
pipeline_white.fit(X_train_white, y_train_white)

print("Logistic Regression pipeline for white wine trained successfully.")
```

#### 4. Evaluate White Wine Model Performance

```python
# Evaluate white wine model on the test set
y_pred_white = pipeline_white.predict(X_test_white)

accuracy_white = accuracy_score(y_test_white, y_pred_white)
print(f"Accuracy of White Wine: {accuracy_white:.4f}")

confusion_white = confusion_matrix(y_test_white, y_pred_white)
print("\nConfusion Matrix of White Wine:\n", confusion_white)

classification_report_white = classification_report(y_test_white, y_pred_white, zero_division=0)
print("\nClassification Report for White Wine:\n", classification_report_white)

macro_f1_white = f1_score(y_test_white, y_pred_white, average='macro', zero_division=0)
print(f"Macro F1-score for White Wine: {macro_f1_white:.4f}")

#print("\n--- White Wine Model Evaluation ---")
#print(f"Accuracy: {accuracy_score(y_test_white, y_pred_white):.4f}")
#print("\nConfusion Matrix:\n", confusion_matrix(y_test_white, y_pred_white))
#print("\nClassification Report (macro F1-score emphasized):\n", classification_report(y_test_white, y_pred_white, zero_division=0))
```

### Comparative Questions Answers

**● Which dataset performed better for Logistic Regression?**

The red wine dataset model achieved a slightly higher accuracy of 0.5906 compared to the white wine model's 0.5490. The macro average F1-score was also marginally better for red wine (0.28 vs 0.24), suggesting better overall performance for red wine.

**● Which quality class looks hardest (lowest recall)?**

For red wine, quality classes 3, 4, and 8 had 0.00 recall, making them the hardest to predict due to very low sample support. In white wine, classes 3, 8, and 9 also showed 0.00 recall for similar reasons. Quality class 4 in white wine also had a very low recall of 0.06.
